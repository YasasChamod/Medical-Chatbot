```python
from langchain.document_loaders import PyPDFLoader, DirectoryLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
# Extract text from PDF files with a safe path resolver
from pathlib import Path

def load_pdf_files(data='data'):
    """Load PDF documents from a `data` folder.
    The function tries the given path first. If it doesn't exist and the notebook
    is running inside a `research/` folder, it will also try the parent directory
    (project root) so `research/trials.ipynb` can load `../data`."""
    p = Path(data)
    # Try the provided path first
    if not p.exists():
        cwd = Path.cwd()
        # If the notebook is running inside `research`, try the repo root's data folder
        if cwd.name == 'research':
            alt = cwd.parent / data
            if alt.exists():
                p = alt
    if not p.exists():
        raise FileNotFoundError(f"Directory not found: '{Path(data)}' (tried '{p.resolve()}')")
    if not p.is_dir():
        raise ValueError(f"Expected a directory, got file: '{p.resolve()}'")
    pdfs = list(p.glob('*.pdf'))
    if not pdfs:
        raise FileNotFoundError(f"No PDF files found in {p.resolve()}. Place .pdf files there and retry.")
    loader = DirectoryLoader(str(p), glob='*.pdf', loader_cls=PyPDFLoader)
    documents = loader.load()
    return documents
```

```python
# Resolve data directory (works when notebook runs from `research/` or project root)
from pathlib import Path

data_dir = Path('data')
if not data_dir.exists() and Path.cwd().name == 'research':
    data_dir = Path.cwd().parent / 'data'
print('Using data dir:', data_dir.resolve())
print('PDF files found:', [p.name for p in data_dir.glob('*.pdf')])
extracted_data = load_pdf_files(str(data_dir))
print(f'Loaded {len(extracted_data)} documents')
```
